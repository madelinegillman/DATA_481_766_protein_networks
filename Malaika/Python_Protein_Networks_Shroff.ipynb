```python
#Trying to visualize a network
import networkx as nx

# Create a graph where nodes are proteins and edges exist if correlation > threshold
def create_network_six(corr_matrix, threshold=0.6):
    G = nx.Graph()
    for col in corr_matrix.columns:
        for row in corr_matrix.index:
            if row != col and abs(corr_matrix.loc[row, col]) > threshold:
                G.add_edge(row, col, weight=corr_matrix.loc[row, col])
    return G

G_high = create_network_six(correlation_matrix_high, threshold)
G_low = create_network_six(correlation_matrix_low, threshold)

plt.figure(figsize=(10, 8))
nx.draw(G_high, with_labels=True, node_color="lightblue", edge_color="gray")
plt.title("Protein Correlation Network - HER2 High")
plt.show()

plt.figure(figsize=(10, 8))
nx.draw(G_low, with_labels=True, node_color="lightcoral", edge_color="gray")
plt.title("Protein Correlation Network - HER2 Low")
plt.show()
#This output looks funny because it is basically taking the whole network and only showing the connections over our threshold
```

```python
# Setting Threshold
specific_threshold = 0.65

# Define a function to get hub proteins based on minimum connections
def get_hub_proteins(corr_matrix, threshold, min_connections=2):
    hub_scores = (corr_matrix.abs() > threshold).sum() - 1  # Exclude self-correlation
    return hub_scores[hub_scores >= min_connections].index

# Define a function to get connected proteins for each hub protein
def get_connected_proteins(corr_matrix, hub_proteins, threshold):
    connected_proteins = {}
    for protein in hub_proteins:
        correlated = corr_matrix.loc[protein][(corr_matrix.loc[protein].abs() > threshold) & 
                                              (corr_matrix.loc[protein].index != protein)].index.tolist()
        connected_proteins[protein] = correlated
    return connected_proteins

# Identify hub proteins for HER2 High and Low at the given threshold
hub_proteins_high_sixfive = get_hub_proteins(correlation_matrix_high, specific_threshold)
hub_proteins_low_sixfive = get_hub_proteins(correlation_matrix_low, specific_threshold)

# Get connections for both HER2 High and Low groups
hub_connections_high_sixfive = get_connected_proteins(correlation_matrix_high, hub_proteins_high_sixfive, specific_threshold)
hub_connections_low_sixfive = get_connected_proteins(correlation_matrix_low, hub_proteins_low_sixfive, specific_threshold)

# Ensure the two hub lists are the same length by padding with empty values
max_length_sixfive = max(len(hub_connections_high_sixfive), len(hub_connections_low_sixfive))

# Pad shorter lists with empty strings to match the max length
high_hubs_list_sixfive = list(hub_connections_high_sixfive.keys()) + [""] * (max_length_sixfive - len(hub_connections_high_sixfive))
high_connections_list_sixfive = [", ".join(proteins) for proteins in hub_connections_high_sixfive.values()] + [""] * (max_length_sixfive - len(hub_connections_high_sixfive))

low_hubs_list_sixfive = list(hub_connections_low_sixfive.keys()) + [""] * (max_length_sixfive - len(hub_connections_low_sixfive))
low_connections_list_sixfive = [", ".join(proteins) for proteins in hub_connections_low_sixfive.values()] + [""] * (max_length_sixfive - len(hub_connections_low_sixfive))

# Create the DataFrame with properly aligned columns
hub_connections_df_fixed_sixfive = pd.DataFrame({
    "HER2 High Hub": high_hubs_list_sixfive,
    "Connected Proteins (HER2 High)": high_connections_list_sixfive,
    "HER2 Low Hub": low_hubs_list_sixfive,
    "Connected Proteins (HER2 Low)": low_connections_list_sixfive
})

# Display in Google Colab
from IPython.display import display
display(hub_connections_df_fixed_sixfive)
```

```python
#She created a function similar to how I did this above
# Function to create a correlation network
def create_network_for_louvain(data, threshold=0.5):
    corr = data.corr()
    G = nx.Graph()
    for col in corr.columns:
        for idx in corr.index:
            if col != idx and abs(corr.at[idx, col]) > threshold:
                G.add_edge(idx, col, weight=abs(corr.at[idx, col])) #We need the negatives to be positives bc of louvain
    return G

def create_network(data, threshold=0.5):
    corr = data.corr()
    G = nx.Graph()
    for col in corr.columns:
        for idx in corr.index:
            if col != idx and abs(corr.at[idx, col]) > threshold:
                G.add_edge(col, idx, weight=corr.at[idx, col])
    return G
    
# Create networks for high and low HER2 groups
high_her2_network = create_network(high_df)
low_her2_network = create_network(low_df)
```

```python
# Function to analyze network properties
def analyze_network(G):
    degree_centrality = nx.degree_centrality(G)
    betweenness_centrality = nx.betweenness_centrality(G)
    clustering_coefficient = nx.clustering(G)
    eigenvector_centrality = nx.eigenvector_centrality(G, max_iter=1000)
    return degree_centrality, betweenness_centrality, clustering_coefficient, eigenvector_centrality

# Analyze networks
high_degree, high_betweenness, high_clustering, high_eigenvector = analyze_network(high_her2_network)
low_degree, low_betweenness, low_clustering, low_eigenvector = analyze_network(low_her2_network)
```

```python
#We will compare the network metrics between High HER2 and Low HER2 groups using the Mann-Whitney U test. 
#This test is appropriate for comparing non-parametric distributions.
def network_metrics_to_df(degree, betweenness, clustering, group):
    df = pd.DataFrame({
        'Protein': list(degree.keys()),
        'Degree Centrality': list(degree.values()),
        'Betweenness Centrality': list(betweenness.values()),
        'Clustering Coefficient': list(clustering.values()),
        'Group': group
    })
    return df

high_metrics_df = network_metrics_to_df(high_degree, high_betweenness, high_clustering, 'High HER2')
low_metrics_df = network_metrics_to_df(low_degree, low_betweenness, low_clustering, 'Low HER2')
```

```python
from scipy.stats import mannwhitneyu

# Combine network metrics for both groups
metrics_df = pd.concat([high_metrics_df, low_metrics_df])

# Function to compare metrics using Mann-Whitney U test
def compare_metrics(df, metric):
    high_values = df[df['Group'] == 'High HER2'][metric]
    low_values = df[df['Group'] == 'Low HER2'][metric]
    stat, p_value = mannwhitneyu(high_values, low_values)
    return p_value

# Perform statistical comparisons for each metric
degree_p_value = compare_metrics(metrics_df, 'Degree Centrality')
betweenness_p_value = compare_metrics(metrics_df, 'Betweenness Centrality')
clustering_p_value = compare_metrics(metrics_df, 'Clustering Coefficient')

# Display the results
print(f"Degree Centrality p-value: {degree_p_value:.4f}")
print(f"Betweenness Centrality p-value: {betweenness_p_value:.4f}")
print(f"Clustering Coefficient p-value: {clustering_p_value:.4f}")
```

```python
#There is a statistically significant difference in degree centrality between high HER2 and low HER2 networks.
#Now lets identify proteins with the higest degree of centrality in each network. 

def print_hub_proteins(graph, name, top_n=5):
    degree_centrality = nx.degree_centrality(graph)
    raw_degrees = dict(graph.degree())  # This gives raw number of connections

    # Combine and sort by centrality
    combined = [(node, degree_centrality[node], raw_degrees[node]) for node in graph.nodes()]
    combined_sorted = sorted(combined, key=lambda x: x[1], reverse=True)

    print(f"Top {top_n} Hub Proteins in {name} Network:")
    for node, centrality, count in combined_sorted[:top_n]:
        print(f"{node}: {centrality:.3f}, {count} connections")
    print()

print_hub_proteins(high_her2_network, "High HER2")
print_hub_proteins(low_her2_network, "Low HER2")
```

```python
#Community detection: Grouping nodes (proteins) into clusters such that nodes within the same group are more connected to each 
#other than to nodes in other groups.
import community as community_louvain

high_partition_for_louvain = community_louvain.best_partition(high_her2_network_for_louvain)
low_partition_for_louvain = community_louvain.best_partition(low_her2_network_for_louvain)
```

```python
from collections import defaultdict

# Group nodes by community
communities_high = defaultdict(list)
for protein, community_id in high_partition_for_louvain.items():
    communities_high[community_id].append(protein)

# Print each community
print("Communities for High HER 2")
for cid, proteins in communities_high.items():
    print(f"Community {cid}: {proteins}")

# Group nodes by community
communities_low = defaultdict(list)
for protein, community_id in low_partition_for_louvain.items():
    communities_low[community_id].append(protein)

# Print each community
print("Communities for Low HER 2")
for cid, proteins in communities_low.items():
    print(f"Community {cid}: {proteins}")
```

```python
# Choose a layout
pos_high = nx.spring_layout(high_her2_network_for_louvain, seed=42)

# Assign colors to communities
community_ids_high = list(set(high_partition_for_louvain.values()))
color_map_high = plt.cm.get_cmap('Set3', len(community_ids_high)) 

# Draw nodes by community
for i, community in enumerate(community_ids_high):
    nodes = [node for node, cid in high_partition_for_louvain.items() if cid == community]
    nx.draw_networkx_nodes(
        high_her2_network_for_louvain,
        pos_high,
        nodelist=nodes,
        node_color=[color_map_high(i)],
        label=f"Community {community}",
        node_size=300
    )
# Draw edges and clean labels
label_pos = {node: (x, y - 0.04) for node, (x, y) in pos_high.items()}
nx.draw_networkx_edges(high_her2_network_for_louvain, pos_high, alpha=0.3)
nx.draw_networkx_labels(
    high_her2_network_for_louvain,
    label_pos,
    font_size=10,
    font_color='black',
    bbox=dict(facecolor='white', edgecolor='none', boxstyle='round,pad=0.2')
)

#Making the legend look better
import matplotlib.patches as mpatches

legend_handles = [
    mpatches.Circle((0, 0), radius=0.1, color=color_map_high(i), label=f"Community {i}")
    for i in range(len(community_ids_high))
]

plt.legend(
    handles=legend_handles,
    loc="lower left",     
    bbox_to_anchor=(0.15, 0.02),  # <-- bump x right from 0 to 0.1
    frameon=True,
    title="Communities",
    labelspacing=0.5,     
    borderpad=0.5   
)

# Final plot touches
plt.title("High HER2 Network (Louvain Communities)")
plt.axis("off")
plt.show()
```

```python
pos_low = nx.spring_layout(low_her2_network_for_louvain, seed=42)  # consistent layout

# Assign colors to communities
community_ids_low = list(set(low_partition_for_louvain.values()))
color_map_low = plt.cm.get_cmap('tab10', len(community_ids_low))  # tab10 gives up to 10 distinct colors

# Draw nodes by community
for i, community in enumerate(community_ids_low):
    nodes = [node for node, cid in low_partition_for_louvain.items() if cid == community]
    nx.draw_networkx_nodes(
        low_her2_network_for_louvain,
        pos_low,
        nodelist=nodes,
        node_color=[color_map_low(i)],
        label=f"Community {community}",
        node_size=300
    )


# Draw edges and labels
label_pos_low = {node: (x, y - 0.05) for node, (x, y) in pos_low.items()}
nx.draw_networkx_edges(low_her2_network_for_louvain, pos_low, alpha=0.3)
nx.draw_networkx_labels(
    low_her2_network_for_louvain,
    label_pos_low,
    font_size=10,
    font_color='black',
    bbox=dict(facecolor='white', edgecolor='none', boxstyle='round,pad=0.2')
)

#Lets Make the Legend look better again
legend_handles_low = [
    mpatches.Circle((0, 0), radius=0.1, color=color_map_low(i), label=f"Community {i}")
    for i in range(len(community_ids_low))
]

plt.legend(
    handles=legend_handles_low,
    loc="upper left",
    bbox_to_anchor=(-0.15, 1),
    frameon=True,
    title="Communities",
    labelspacing=0.5,     
    borderpad=0.5   
)

plt.title("Low HER2 Network (Louvain Communities)")
plt.axis("off")
plt.show()
```
